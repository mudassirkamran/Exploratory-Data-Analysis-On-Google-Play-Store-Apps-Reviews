```python
# To change the data type of the Size column we must first remove the letter (M/k/G)

def fix_size(s):
  try:
    if s.endswith('M'):
      return float(s[:-1])
    elif s.endswith('k'):
      return (float(s[:-1]) / 1024)
    elif s == '1,000+':
      return 1.0
    else:
      return np.nan
  except:
    return np.nan


# def fix_size(s):
#   try:
#     if s.endswith('M'):
#       return float(s[:-1])
#     elif s.endswith('k'):
#       return (float(s[:-1]) / 1024)
#     elif s == '1,000+':
#       return 1.0
#     else:
#       return np.nan
#   except:
#       return np.nan
```

```python
#Testing the funtion..

print(fix_size('98M'))
print(fix_size('900k'))
print(fix_size('1,000+'))
```

```python
# changing the values of the size column into numerical-mb.
df['Size'] = df['Size'].apply(fix_size)
```

```python
#Custom function to remove the $ sign and changing the data type of the prize column
def fix_price(p):
  try:
    if p.startswith('$'):
      return float(p[1:])
    else:
      return 0.0
  except:
    np.nan
```

```python
#apply on column

df['Price'] = df['Price'].apply(fix_price)
```

```python
#There are some invalid values in the column:
# Varies with device
# 4.4W and up

def correction(y):
  try:
    if y == 'Varies with device':
      return np.nan
    elif y == '4.4W and up':
      return y.replace('W','')
    else:
      return y
  except:
      return y
```

```python
df['Android Ver'] = df['Android Ver'].apply(correction)
```
